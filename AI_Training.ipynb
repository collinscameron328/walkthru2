```python
#id first_training
#caption Results from the first training
# CLICK ME
from fastai.vision.all import *
path = "/home/ccollins328/git/walkthru2/images/"
print(path)
print(get_image_files(path))
files = get_image_files(path)

def is_cat(x): 
    return x[0].isupper()

print(files[1])
dls = ImageDataLoaders.from_name_func(
    path, get_image_files(path), valid_pct=0.3, seed=26,
    label_func=is_cat, item_tfms=Resize(200))

learn = vision_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(5)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(5, nrows=1)
```

```python
uploader1 = SimpleNamespace(data = ['images/test_image.jpg'])
```

```python
img1 = PILImage.create(uploader1.data[0])
img1.to_thumb(192)
```

```python
is_cat,_,probs = learn.predict(img1)
print(f"Is this a cat?: {is_cat}.")
print(f"Probability it's a cat: {probs[1].item():.6f}")
```
